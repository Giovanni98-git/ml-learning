```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns # permet de la visualisation de donnnées poussés en écrivant une ligne de code 
```

```python
iris = pd.read_csv('Iris.csv' )
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.pairplot(titanic)
```

```python
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.displot(titanic['fare'])
```

```python
sns.jointplot(x='age',y='fare',data=titanic, kind='kde')
```
