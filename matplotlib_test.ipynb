```python
import numpy as np 
```

```python
x = np.linspace(0,2,10)
y = x**2
print(x)
```

# 1) Pyplot

```python
import matplotlib.pyplot as plt
```

```python
# Pour tracer un graphique
plt.plot(x,y)
plt.show()
```

```python
# Pour tracer un nuage de point
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y,c='red') # c permet de choisir la couleur de la courbe
plt.show()
```

```python
plt.plot(x,y,c='red', lw=3) # lw permet de choisir l'épaisseur de la courbe
plt.show()
```

```python
plt.plot(x,y,c='red', lw=3, ls='--') # lw permet de choisir le corps de la courbe
plt.show()
```

```python
# Le cycle de vie d'une figure 
# Pour tracer une figure ou un schema 
# il est conseillé de commencer par la fonction figure de plt
# figsize=(h,w) permet de choisir la taille de la figure
# plt.figure(figsize=(12,8)) # Début de la figure
plt.figure()
plt.plot(x,y,label='quadratique') # ajout de la légende
plt.plot(x,x**3, label='cubique') # ajout d'une seconde figure
plt.title('figure 1') # ajout d'un titre
plt.xlabel('axe x') # ajout d'un nom pour l'axe x
plt.ylabel('axe y') # ajout d'un nom pour l'axe y
plt.legend()
plt.show() # affiche la figure
plt.savefig('figure.png') # sauvegarge l'image de la courbe
```

# 2) Subplot 
Elle nous permet de générer une grille de différentes graphiques sur notre figure

```python
plt.subplot(2,1,1) # trace un graphique
plt.plot(x,y,c='red')
plt.subplot(2,1,2)
plt.plot(x,y,c='blue')
```

# OOP
Permet de tracer des graphique avec la programmation orientée objet et avec elle on a plus de possibilité mais elle n'es pas souvent utilisé

```python
fig , ax = plt.subplots(2,1,sharex=True)
# fig est un objet et ax est un tableau ndarray qui contient des objets
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
plt.show()
```

## Exercice 

```python
dataset = {f"experience {i}": np.random.randn(100) for i in range(4)}
dataset
def graphique(dataset):
    for key in dataset.keys():
        x = [i for i in range(len(dataset[key]))]
        plt.figure(figsize=(15,8))
        plt.subplot(2,2,1)
        plt.plot(x,dataset[key])
        plt.xlabel(key)
        plt.show()
        
graphique(dataset)
```
