## NaN Corrections

```python
A  = np.random.randn(5,5)
A[0,2] = np.nan
A[4,3] = np.nan
A
```

```python
np.nanmean(A)
```

```python
np.isnan(A).sum()/A.size
```

```python
A[np.isnan(A)] = 0
A
```
